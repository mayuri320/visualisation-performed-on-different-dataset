# Q.3: Visualization – Movies

```python
#reading csv file
df3 = pd.read_csv("https://raw.githubusercontent.com/flaviovdf/evcomp2018/master/data/movies_by_year.csv")
df3
```

```python
#Q.3.1) Plot ('Year' vs  'Number of Movies')
df3.plot.line(x='Year', y='Number of Movies')
```

```python
#Q.3.2) Plot ('Year' vs  'Number of Movies') – 21st century

year2000 = df3[df3["Year"] >=2000].dropna() #take only years greater than 2000 
year2000.plot.line(x='Year', y='Number of Movies')
```

```python
#Q.3.3) Analyse reason for sharp drop in the number of movies released in 2009
df3[(df3['Year'] >= 2008) & (df3['Year'] <= 2010) ]
```

When analysing the three consecutive years 2008, 2009 and 2010, it can be seen that the AVATAR movie is the number one movie in the year 2009 and the high increase in the box office receipts might due the AVATAR movie which the film releases might have thought its best to release their films after the 'hype' and rush towards AVATAR.

# Q.4: Visualization



```python
df4 = pd.read_csv("https://github.com/flaviovdf/evcomp2018/raw/master/data/nc-est2015-agesex-res.csv")
df4
```

```python
#Q.4.1) 
#['SEX', 'AGE', 'POPESTIMATE2010', 'POPESTIMATE2014'] 

df4_new = df4[['SEX', 'AGE', 'POPESTIMATE2010', 'POPESTIMATE2014']]
df4_new
```

```python
#Q.4.2) 

df4_new = df4_new.rename(columns={'POPESTIMATE2010': '2010','POPESTIMATE2014': '2014'})
df4_new
```

```python
#Q.4.3) 
age_18=df4_new[(df4_new['AGE']<=18 )]
age_18
```

```python
#taking chidren with sex=0
age_18=age_18[(age_18['SEX']==0 )].drop("SEX",axis=1)
age_18
```

```python
#Q.4.4)  
#the years 2010 and 2014.

age=age_18.drop("AGE",axis=1)
ax=age.plot()
ax.set_xlabel("AGE")
```

```python
#Data: usa_ca_2014.csv
df6 = pd.read_csv('https://raw.githubusercontent.com/flaviovdf/evcomp2018/master/data/usa_ca_2014.csv')
df6
```

```python
#Q.6.1) 
#because all the entries are proportions and are therefore on the same scale. Visualize the comparisons 
#by drawing multiple bar charts on the same axes.

df6.plot.barh(x='Ethnicity').invert_yaxis()

#bar"h" to plot horizontally
```

```python
#Q.6.2)
df6[['Ethnicity', 'USA All', 'CA All']].plot.barh(x='Ethnicity').invert_yaxis()
```

```python
#Q.6.3) 
#likely to be greater in future years. Among Californian children, 50% are in the Hispanic category (show).

df6[['Ethnicity','CA All','CA Children']].plot.barh(x='Ethnicity').invert_yaxis()
```
